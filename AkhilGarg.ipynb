```python
# Numerical and Data Analysis
import numpy as np
import pandas as pd

# Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Extra - To Suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
## Reading loan data csv file from source
df=pd.read_csv("D:/datasets//loan.csv")
```

```python
# Listing top 5 records from data frame
df.head()
```

```python
## Display total no. of rows and columns in data frame
df.shape
```

```python
## Display data types of different columns
df.info()
```

```python
#List all null columns
len(df.columns[df.isna().all()].tolist())
```

```python
#Drop all null columns
df.dropna(how='all', axis=1, inplace=True)

##Null columns count : 54
```

```python
## List all columns
df.info()
## Non null columns 57
```

```python
## check null percentage
100*df.isnull().mean()
```

```python
# drop column having high (30%+) null values
df1=df.drop(["mths_since_last_delinq","next_pymnt_d","mths_since_last_record","desc"],axis=1)
```

```python
## check null percentage
100*df1.isnull().mean()
```

```python
## drop all null values from all rows
df2=df1.dropna(axis=0,how='any')
```

```python
## check null percentage
100*df2.isnull().mean()
```

```python
## clean emp length column. Remove years and convert to int data type
df2["emp_length"]=df2["emp_length"].str.replace('+',"")
df2["emp_length"]=df2["emp_length"].str.replace('years',"")
df2["emp_length"]=df2["emp_length"].str.replace('year',"")
df2["emp_length"]=df2["emp_length"].str.replace('<',"")
df2["emp_length"]=df2["emp_length"].str.strip()
df2["emp_length"]=df2["emp_length"].astype('int32')
```

```python
## clean interest rate column. Remove '%' and convert to float data types
df2["int_rate"]=df2["int_rate"].str.replace('%',"")
df2["int_rate"]=df2["int_rate"].str.strip()
df2["int_rate"]=df2["int_rate"].astype('float')
```

```python
## clean term column. Remove 'months' and convert to int data types
df2["term"]=df2["term"].str.replace('months',"")
df2["term"]=df2["term"].str.strip()
df2["term"]=df2["term"].astype('int')
```

```python
## clean revol util column. Remove '%' and convert to float data types
df2["revol_util"]=df2["revol_util"].str.replace('%',"")
df2["revol_util"]=df2["revol_util"].str.strip()
df2["revol_util"]=df2["revol_util"].astype('float')
```

```python
## Create derived column 'IncInstallmentRatio' income vs installment ratio
df2["IncInstallmentRatio"] = (df2["installment"]*100)/ df2["annual_inc"]
```

```python
## Create derived column 'grade_Numbers'. Created based on grades to get the numeric values for bivariate analysis
df2["grade_Numbers"] = df2["grade"].replace({"A":1,"B":2,"C":3,"D":4,"E":5,"F":6,"G":7})
```

```python
## Created derived column 'addr_state_number' based on state. Gave number to each state to get the numeric values for bivariate analysis
df2["addr_state_number"] = df2["addr_state"].replace({'GA':1, 'CA':2, 'OR':3, 'AZ':4, 'NC':5, 'TX':6, 'VA':7, 'IL':8, 'MO':9, 'CT':10, 'UT':11,
       'FL':12, 'PA':13, 'MN':14, 'NJ':15, 'KY':16, 'OH':17, 'SC':18, 'RI':19, 'LA':20, 'MA':21, 'NY':22,
       'WA':23, 'WI':24, 'AL':25, 'CO':26, 'KS':27, 'NV':28, 'AK':29, 'MD':30, 'WV':31, 'VT':32, 'MI':33,
       'DC':34, 'NH':35, 'AR':36, 'NM':37, 'MT':38, 'HI':39, 'WY':40, 'OK':41, 'SD':42, 'DE':43, 'MS':44,
       'TN':45, 'IA':46, 'NE':47, 'ID':48})
```

```python
## Create derived column 'HomeOwnership_Numbers' based on home ownership to get the numeric values for bivariate analysis
df2["HomeOwnership_Numbers"] = df2["home_ownership"].replace({"RENT":1,"OWN":2,"MORTGAGE":3,"OTHER":4})
df2["HomeOwnership_Numbers"] = df2["HomeOwnership_Numbers"].astype('int')
```

# Univariate Analysis

```python
## Total count and percentage of loan customer by their loan status (Fully Paid/Charged Off/Current)
ax = sns.countplot(x=df2['loan_status'],
                   order=df2['loan_status'].value_counts(ascending=False).index)

abs_values = df2['loan_status'].value_counts(ascending=False)
rel_values = df2['loan_status'].value_counts(ascending=False, normalize=True).values * 100
lbls = [f'{p[0]} ({p[1]:.0f}%)' for p in zip(abs_values, rel_values)]

ax.bar_label(container=ax.containers[0], labels=lbls)
ax.set_title("Loan Status")

plt.show()
```

```python
## Total count and percentage of loan customer by Grade (A/B/C/D/E/F/G)
ax = sns.countplot(x=df2['grade'],
                   order=df2['grade'].value_counts(ascending=False).index);

abs_values = df2['grade'].value_counts(ascending=False)
rel_values = df2['grade'].value_counts(ascending=False, normalize=True).values * 100
lbls = [f'{p[0]} ({p[1]:.0f}%)' for p in zip(abs_values, rel_values)]

ax.bar_label(container=ax.containers[0], labels=lbls)
ax.set_title("Grade")
plt.show()
```
